```python
STOP_HERE
%%js
// Draw The Grid ... and allocate geometry for the layout, the drag rect, and the interaction layer
var grid =  '<svg id="placer" width="1000" height="1000">'
grid += '<rect x="0" y="0" width="1000" height="1000" fill="#000000"/>'
for (let i=0;i<40;i++) {
  grid += '<line x1="'+i*25+'" y1="'+0+'" x2="'+i*25+'" y2="'+1000+'" stroke="#303030" />';
  grid += '<line y1="'+i*25+'" x1="'+0+'" y2="'+i*25+'" x2="'+1000+'" stroke="#303030" />';
};
for (let i=0;i<100;i++) {
  grid += '<rect id="r'+i+'" x="-10" y="-10" width="5" height="5" stroke="#0000ff" stroke-width="3" opacity="0.5"/>';
  grid += '<text id="t'+i+'" x="-10" y="-10" fill="#ffffff">r'+i+'</text>'; 
}
grid += '<rect id="drag" x="-10" y="-10" width="5" height="5" fill="none" stroke="#ff0000" stroke-width="1"/>';
grid += '<rect id="interact" x="0" y="0" width="1000" height="1000" fill="#000000" opacity="0.1"/>';
grid += '</svg>';
document.write(grid);

//
// MyRect - class for storing the layout coordinates
//
class MyRect {
	constructor(el_str, xa_i, ya_i, xb_i, yb_i) {
  	this.el_str = el_str; this.xa_i = xa_i; this.ya_i = ya_i; this.xb_i = xb_i; this.yb_i = yb_i;
  }
  contains(x,y) {
  	if ((x >= 25*this.xa_i) && (x < 25*this.xb_i) && (y >= 25*this.ya_i) && (y < 25*this.yb_i))
  		return true;
  }
}

//
// Map For Placed Layout Rectangles
//
const rects = new Map();

// Drag coordinates & flag
var   x0 = -5, y0 = -5, x1 = 10, y1 = 10, drag_op = false;

//
// drawDragOp() - draw the drag operation
//
function drawDragOp() {
	if (x0 < x1) { xa = x0; xb = x1; } else { xa = x1; xb = x0; }
  if (y0 < y1) { ya = y0; yb = y1; } else { ya = y1; yb = y0; }
  xa = Math.floor(25*(Math.floor(xa/25.0))); ya = Math.floor(25*(Math.floor(ya/25.0)));
  xb = Math.ceil (25*(Math.ceil (xb/25.0))); yb = Math.ceil (25*(Math.ceil (yb/25.0)));
  el = document.getElementById('drag')
  el.setAttribute('x',      xa);         el.setAttribute('y',      ya);
  el.setAttribute('width',  (xb - xa));  el.setAttribute('height', (yb - ya));
}

//
// resetDragOp() - reset the drag operation
//
function resetDragOp() {
  el = document.getElementById('drag');
  el.setAttribute('x',      -10); el.setAttribute('y',      -10);
  el.setAttribute('width',    5); el.setAttribute('height',   5);
}

//
// mouseDown() - process mouse down event
// - if the mouse started in an existing layout rectange, delete it
// - else start a drag op
//
function mouseDown(event) {
  remove_happened = false;
	for (const key of rects.keys()) {
  	if (rects.get(key).contains(event.x, event.y)) {
  		el = document.getElementById(key);
  		el.setAttribute('x',      -10); el.setAttribute('y',      -10);
  		el.setAttribute('width',    5); el.setAttribute('height',   5);
      el = document.getElementById('t'+key.substring(1));
      el.setAttribute('x',      -10); el.setAttribute('y',      -10);
    	rects.delete(key);
      remove_happened = true;
    }
	}
  if (remove_happened == false) {
			drag_op = true; x0 = x1 = event.x; y0 = y1 = event.y; drawDragOp();
    }
}

//
// mouseUp() - process the mouse up event
// - if there's a drag op, create a layout rectange & display it
//
function mouseUp(event) {
	if (drag_op) {
  	x1 = event.x; y1 = event.y; drag_op = false; resetDragOp();
 	  el_str = t_str = null;
  	for (i=0;i<100;i++) {
  		el_str = 'r' + i; t_str = 't' + i
    	if (rects.has(el_str) == false) break;
  	}
  	if (el_str != null) {
    	xa_i = Math.floor(xa/25.0); ya_i = Math.floor(ya/25.0);
    	xb_i = Math.ceil (xb/25.0); yb_i = Math.ceil (yb/25.0);
    	xa = Math.floor(25*(Math.floor(xa/25.0))); ya = Math.floor(25*(Math.floor(ya/25.0)));
    	xb = Math.ceil (25*(Math.ceil (xb/25.0))); yb = Math.ceil (25*(Math.ceil (yb/25.0)));
    	el = document.getElementById(el_str);
    	el.setAttribute('x',      xa);         el.setAttribute('y',      ya);
    	el.setAttribute('width',  (xb - xa));  el.setAttribute('height', (yb - ya));
    	el = document.getElementById(t_str);
    	el.setAttribute('x',      xa+5);       el.setAttribute('y',      ya+20);
    	rects.set(el_str, new MyRect(el_str, xa_i, ya_i, xb_i, yb_i))
    }
  }
}

//
// mouseMove() - process the mouse move event
// - if there's a drag op, update the drag rectangle
//
function mouseMove(event) {
	if (drag_op) { x1 = event.x; y1 = event.y; drawDragOp(); }
}

// Assign Event Listeners
document.getElementById('interact').addEventListener('mousedown',  mouseDown)
document.getElementById('interact').addEventListener('mouseup',    mouseUp)
document.getElementById('interact').addEventListener('mousemove',  mouseMove)
```

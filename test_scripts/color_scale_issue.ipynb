```python
import polars as pl
import rtsvg
rt = rtsvg.RACETrack()

_rt_params_  = {'x_field':'xaxis', 'y_field':'yaxis', 'color_by':'cfield', 'count_by':'cfield', 'color_magnitude':'linear', 'dot_size':'small'}

_lu_ = {'xaxis':[], 'yaxis':[], 'cfield':[]}
for xi in range(3000): 
    for yi in range(3000): _lu_['xaxis'].append(xi-10_000), _lu_['yaxis'].append(yi+20_000), _lu_['cfield'].append(2*xi-yi)
df_17 = pl.DataFrame(_lu_)

_tiles_   = []
for _sz_ in [64, 128, 256, 512]: _tiles_.append(rt.xy(df_17, w=_sz_, h=_sz_, align_pixels=True, **_rt_params_))
rt.tile(_tiles_)
```

```python
rt.co_mgr.spectrum(0.5, 0, 200)
_svg_ = ['<svg x="0" y="0" width="1024" height="64">']
for x in range(1024): _svg_.append(f'<rect x="{x}" y="0" width="1" height="64" fill="{rt.co_mgr.spectrum(x, 0, 1024-1)}" />')
_svg_.append('</svg>')
rt.tile([''.join(_svg_)])
```
